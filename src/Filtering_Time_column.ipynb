```python
attacks_filtered = pd.read_csv("../data/Attacks_filtered3.csv")

attacks_filtered
```

We are going to focus in column "Time", trying to clean as many data as possible

```python
print(list(attacks_filtered["Time"]))
```

```python
# %run Functions.ipynb 
# filter_time(attacks_filtered["Time"])
for i in attacks_filtered["Time"]:
        i = str(i)

        if "06" in i or "07"  in i or "08" in i or "09" in i or "10" in i or "11" in i or "morning" in i or "Morning" in i or "A.M" in i or "AM" in i or "dawn" in i or "Dawn"in i:
            attacks_filtered["Time"] = attacks_filtered["Time"].replace(i, "Morning")

        elif "12" in i or "13"in i or "14"in i or "15"in i or "16"in i or "17"in i or "afternoon"in i or "lunch"in i or "noon"in i or "Noon"in i or "Midday"in i or "P.M." in i:      
            attacks_filtered["Time"] = attacks_filtered["Time"].replace(i, "Afternoon")

        elif "18"in i or "19"in i or "20"in i or "21"in i or "Sunset"in i or "dusk"in i or "Dusk"in i or "evening" in i:
            attacks_filtered["Time"] = attacks_filtered["Time"].replace(i, "Evening")

        elif "22"in i or "23"in i or "24"in i or "01"in i or "02"in i or "03"in i or "04"in i or "05"in i or "Night"in i or "night"in i:
            attacks_filtered["Time"] = attacks_filtered["Time"].replace(i, "Nightime")
            
#         else:
#              attacks_filtered["Time"] = attacks_filtered["Time"].replace(i, "Random, or not clear time references")
            
        

    
```
